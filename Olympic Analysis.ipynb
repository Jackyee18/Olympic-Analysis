# Find the history of each sport by finding the first year, last year and total years that sport played in the Olympics

# Olympics Gender Ratio


Find the gender ratio between the number of men and women who participated in each Olympics.
Output the Olympics name along with the corresponding number of men, women, and the gender ratio. If there are Olympics with no women, output a NULL instead of a ratio.


```python
gender=q[q.Sex=='M'].groupby('Games')['ID'].count().reset_index().merge(
    q[q.Sex=='F'].groupby('Games')['ID'].count().reset_index(),on='Games').rename(columns=
                                                                                  {'ID_x':'Total Male','ID_y':'Total Female'})
gender.head()
```

```python
gender['Male_pct']=(gender['Total Male']/(gender['Total Male']+gender['Total Female'])*100).round(2)
gender['Female_pct']=(gender['Total Female']/(gender['Total Male']+gender['Total Female'])*100).round(2)
```

# Find how the average male height changed between each Olympics from 1896 to 2016

Find how the average male height changed between each Olympics from 1896 to 2016.
Output the Olympics year, average height, previous average height, and the corresponding average height difference.
Order records by the year in ascending order.


If avg height is not found, assume that the average height of an athlete is 172.73.

```python
avg_height=q[(q.Sex=='M')&(q.Year>=1896)&(q.Year<=2016)].groupby('Year')['Height'].mean().reset_index().sort_values('Year')
avg_height.head()
```

```python
avg_height['previous_height']=avg_height.Height.shift(1)
```

```python
avg_height['height_diff']=avg_height.Height.diff()
```

# Olympic Medals By Chinese Athletes


Find the number of medals earned in each category by Chinese athletes from the 2000 to 2016 summer Olympics. For each medal category, calculate the number of medals for each olympic games along with the total number of medals across all years. Sort records by total medals in descending order.

```python
cn_medal=q[(q.Year<=2016)&(q.Year>=2000)&(q.Season=='Summer')&(q.NOC=='CHN')].groupby('Games').Medal.value_counts().reset_index(name='count')
cn_medal.head()
```

```python
cn_medal1=pd.pivot(cn_medal,index='Games',columns='Medal',values='count').reset_index()
cn_medal1
```

```python
cn_medal1['Total']=cn_medal1['Bronze']+cn_medal1['Bronze']+cn_medal1['Bronze']
cn_medal1
```

# Top Teams In The Rio De Janeiro 2016 Olympics

Find the top 3 medal-winning teams by counting the total number of medals for each event in the Rio De Janeiro 2016 olympics. In case there is a tie, order the countries by name in ascending order. Output the event name along with the top 3 teams as the 'gold team', 'silver team', and 'bronze team', with the team name and the total medals under each column in format "{team} with {number of medals} medals". Replace NULLs with "No Team" string.

```python
olym_2016_total= q[q.Medal.str.contains('Gold|Silver|Bronze',na=False)&(q.Year==2016)].groupby(
    'region')['Medal'].value_counts().reset_index(name='total')
olym_2016_total
```

```python
olym_2016_total=pd.pivot(olym_2016_total,columns='Medal',values='total',index='region').reset_index().fillna(0)
olym_2016_total
```

```python
olym_2016_total['Total']=olym_2016_total.Bronze+olym_2016_total.Gold+olym_2016_total.Silver
```

```python
olym_2016_total.loc[:,['region','Total','Gold','Silver','Bronze']].sort_values('Total',ascending=False)
```
